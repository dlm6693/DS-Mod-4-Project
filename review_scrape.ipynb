```python
def review_scrape(url_list):    
    options = Options()
    options.headless = True
    browser = webdriver.Chrome(options=options) 
    reviews = []
    for url in tq(url_list):
        browser.get(url)
        try:
            browser.find_element_by_xpath("//a[@id='bookDataBoxShow']").click()
            container = browser.find_element_by_id('bookDataBox')
            isbn = container.find_elements_by_class_name('infoBoxRowItem')[1].text
        except:
            isbn = ''
        for x in tq(range(2,11)):
            for x in tq(browser.find_elements_by_css_selector('.friendReviews.elementListBrown')):
                review = {}
                try:
                    name = x.find_element_by_class_name('user').get_attribute('name')
                    review['name'] = name
                except:
                    review['name'] = ''
                try:
                    id_ = x.find_element_by_class_name('user').get_attribute("href").split('/')[-1].split('-')[0]
                    review['id'] = id_
                except:
                    review['id'] = ''
                try:
                    rating = x.find_element_by_css_selector('.staticStars.notranslate').get_attribute('title')
                    review['rating'] = rating
                except:
                    review['rating'] = ''
                try:
                    likes = x.find_element_by_class_name('likesCount').text
                    review['likes'] = likes
                except:
                    review['likes'] = ''
                review['url'] = url
                review['isbn'] = isbn
                reviews.append(review)
            try:
                time.sleep(1.5)
                browser.find_element_by_partial_link_text('next').click()
            except:
                pass
    browser.quit()
    return pd.DataFrame(reviews)
```

```python
reviews = review_scrape(urls)
reviews.to_csv('reviews.csv')
```
